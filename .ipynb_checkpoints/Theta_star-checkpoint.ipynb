```python
import heapq
import math
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib
%matplotlib inline
#set up matplotlib
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
plt.ion()
plt.figure(figsize=(12, 12))
MAX_INT = 9999.0

class GridMap:
    def __init__(self, ox, oy, resolution):
        self.min_x = None
        self.min_y = None
        self.max_x = None
        self.max_y = None
        self.x_width = None
        self.y_width = None
        self.obstacle_map = None
        
        self.resolution = resolution
        self.calc_obstacle_map(ox, oy)
        
    def calc_obstacle_map(self, ox, oy):

        self.min_x = round(min(ox))
        self.min_y = round(min(oy))
        self.max_x = round(max(ox))
        self.max_y = round(max(oy))
        print("min_x:", self.min_x)
        print("min_y:", self.min_y)
        print("max_x:", self.max_x)
        print("max_y:", self.max_y)

        self.x_width = round((self.max_x - self.min_x) / self.resolution)+1
        self.y_width = round((self.max_y - self.min_y) / self.resolution)+1
        print("x_width:", self.x_width)
        print("y_width:", self.y_width)

        # obstacle map generation
        self.obstacle_map = np.zeros((self.x_width, self.y_width))
        for x, y in zip(ox, oy):
            ix, iy = self.calc_xy_index(x, self.min_x), self.calc_xy_index(y, self.min_y)
            if 0<=ix<self.x_width and 0<=iy<self.y_width:
                self.obstacle_map[ix, iy] = 10.0  
        

    def calc_position(self, index, minp):
        pos = index * self.resolution + minp
        return pos
    
    def get_cx1x2_by_two_nodes(self, x1, y1, x2, y2):
        if x1 == x2 and y1 != y2:
            iyc = iyb = min(y1, y2)
            ixc = max(x1 - 1, 0)
            ixb = min(x1, self.x_width-1)
            return min(self.get_grid_cost(ixc, iyc), self.get_grid_cost(ixb, iyb))*self.resolution
            
        if x1 != x2 and y1 == y2:
            ixc = ixb = min(x1, x2)
            iyc = max(y1 - 1, 0)
            iyb = min(y1, self.y_width-1)
            return min(self.get_grid_cost(ixc, iyc), self.get_grid_cost(ixb, iyb))*self.resolution 
            
        if x1 != x2 and y1 != y2:
            ixc = min(x1, x2)
            iyc = min(y1, y2)
            return self.get_grid_cost(ixc, iyc)*self.resolution*1.414
            
    def get_grid_cost(self, ix, iy):
        if abs(self.obstacle_map[ix, iy]-10.0) < 0.01:
            return MAX_INT  
        else:
            return 1.0            

    def calc_xy_index(self, position, minp):
        return round((position - minp) / self.resolution)
    
    def calc_unique_index(self, ix, iy):
        return iy*self.x_width + ix

    def verify_node(self, ix, iy):
        if ix >= self.x_width or ix < 0 or iy >= self.y_width or iy < 0:
            return False      
        return True

sx = -5.0  # [m]
sy = -5.0  # [m]
gx = 50.0  # [m]
gy = 50.0  # [m]
grid_size = 5.0  # [m]
robot_radius = 1.0  # [m]

# set obstacle positions
ox, oy = [], []
for i in range(-10, 60):
    ox.append(i)
    oy.append(-10.0)
for i in range(-10, 60):
    ox.append(60.0)
    oy.append(i)
for i in range(-10, 61):
    ox.append(i)
    oy.append(60.0)
for i in range(-10, 61):
    ox.append(-10.0)
    oy.append(i)
# for i in range(-10, 40):
#     ox.append(20.0)
#     oy.append(i)
for i in range(20, 40):
    ox.append(40.0)
    oy.append(60.0 - i)    
    
for i in range(20, 40):
    oy.append(40.0)
    ox.append(i)      
grid_map = GridMap(ox, oy, grid_size)




class Node:
    def __init__(self, ix, iy, C=0.0, P=None):
        self.ix = ix
        self.iy = iy
        self.C = C
        self.P = P
        
action_set = {'UP':[0, 1], 'DOWN':[0, -1], 'LEFT':[-1, 0], 'RIGHT':[1, 0], 'UL':[-1, 1], 'UR':[1, 1], 'DL':[-1, -1], 'DR':[1, -1]}

class Astar(object):
    def __init__(self, grid_map, start_node, end_node):
        self.start_node = start_node
        self.end_node = end_node
        self.grid_map = grid_map
        self.start_uid = self.grid_map.calc_unique_index(start_node.ix, start_node.iy)
        self.end_uid = self.grid_map.calc_unique_index(end_node.ix, end_node.iy)
        self.Closed_set = {}
        self.Open_set = {}
        self.Q = []
        self.Open_set[self.start_uid] = start_node
        heapq.heappush(self.Q, (start_node.C+self.get_heuristic(start_node), 
                           self.grid_map.calc_unique_index(start_node.ix, start_node.iy))) 
        
    def get_heuristic(self, node):
        dx = node.ix - self.end_node.ix
        dy = node.iy - self.end_node.iy
        return np.sqrt(dx*dx+dy*dy)*self.grid_map.resolution  
        

    def get_neighbors(self, current):
        for a in action_set:
            ix, iy = current.ix+action_set[a][0], current.iy+action_set[a][1]
            if self.grid_map.verify_node(ix, iy):
                cx1x2 = self.grid_map.get_cx1x2_by_two_nodes(current.ix, current.iy, ix, iy)                   
                yield (ix, iy, cx1x2)  
                
    def compute_shortest_path(self):
        while True:
            if not self.Open_set:
                print("A star Algorithm Finished.")
                return [], [], -1

            _, c_id = heapq.heappop(self.Q)

            if c_id in self.Open_set:
                current = self.Open_set.pop(c_id)
                self.Closed_set[c_id] = current
                if current.ix == end_node.ix and current.iy == end_node.iy:
                    return self.ShortestPathFinder()
            else:
                continue
                
            for ix, iy, cx1x2 in self.get_neighbors(current):
                neighbor_uid = self.grid_map.calc_unique_index(ix, iy)
                if neighbor_uid in self.Closed_set: continue
                if neighbor_uid not in self.Open_set:
                    self.Open_set[neighbor_uid] = Node(ix, iy, current.C+cx1x2, c_id)
                    heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid)) 
                elif self.Open_set[neighbor_uid].C > (current.C+cx1x2):
                    self.Open_set[neighbor_uid].C = current.C+cx1x2
                    self.Open_set[neighbor_uid].P = c_id
                    heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid))  
                     
                    
    def ShortestPathFinder(self):
        SPath_x = [self.Closed_set[self.end_uid].ix]
        SPath_y = [self.Closed_set[self.end_uid].iy]      
        SPath_dist = self.Closed_set[self.end_uid].C*self.grid_map.resolution
        v_index = self.Closed_set[self.end_uid].P
        while v_index: 
            SPath_x.append(self.Closed_set[v_index].ix)
            SPath_y.append(self.Closed_set[v_index].iy)
            v_index = self.Closed_set[v_index].P
        return list(reversed(SPath_x)), list(reversed(SPath_y)), SPath_dist   
    
    def LineOfSight(self, ix1, iy1, ix2, iy2):            
        xL = abs(ix2 - ix1)
        yL = abs(iy2 - iy1)
        sign_x, sign_y = 1, 1
        if ix2 < ix1: sign_x = -1
        if iy2 < iy1: sign_y = -1
            
        if ix1 == ix2 and iy1 != iy2:
            for i in range(yL):
                iyc = iyb = min(iy1 + sign_y*i, iy1 + sign_y*(i+1))
                ixc = max(ix1 - 1, 0)
                ixb = min(ix1, self.grid_map.x_width-1)
                if min(self.grid_map.get_grid_cost(ixc, iyc), self.grid_map.get_grid_cost(ixb, iyb)) > 1.0:
                    return False
            return True
            
        if ix1 != ix2 and iy1 == iy2:
            for i in range(xL):
                ixc = ixb = min(ix1 + sign_x*i, ix1 + sign_x*(i+1))
                iyc = max(iy1 - 1, 0)
                iyb = min(iy1, self.grid_map.y_width-1)
                if min(self.grid_map.get_grid_cost(ixc, iyc), self.grid_map.get_grid_cost(ixb, iyb)) > 1.0:
                    return False
            return True
            
        tan_theta = yL/xL
        for i in range(xL):
            ixt = ix1 + sign_x*i
            iyt = iy1 + sign_y*int(i*tan_theta)
            if self.grid_map.get_grid_cost(ixt, iyt) > 1.0:
                return False  
    def PostSmoothPath(self, path_x, path_y):
        k = 0
        ps_path_x, ps_path_y = [path_x[0]], [path_y[0]]
        for i in range(1, len(path_x)-1):
            if self.LineOfSight(ps_path_x[k], ps_path_y[k], path_x[i+1], path_y[i+1]) == False:
                k += 1
                ps_path_x.append(path_x[i])
                ps_path_y.append(path_y[i])
        
        ps_path_x.append(path_x[-1])
        ps_path_y.append(path_y[-1])
        return ps_path_x, ps_path_y
        
       

start_node = Node(2, 2)
end_node = Node(12, 12)
plt.imshow(grid_map.obstacle_map)
import time
t0 = time.time()
planner = Astar(grid_map, start_node, end_node)
path_x, path_y, path_dist = planner.compute_shortest_path()
ps_path_x, ps_path_y = planner.PostSmoothPath(path_x, path_y)
print("time cost is:", time.time()-t0)
grid_map.obstacle_map[end_node.ix, end_node.iy] = 4
for ix, iy in zip(path_x, path_y):
    grid_map.obstacle_map[ix, iy] = 4  

plt.scatter(2-0.5, 2-0.5, color='green', linewidth=12.0)
plt.scatter(12-0.5, 12-0.5, color='red', linewidth=12.0)
# plt grid
for i in range(grid_map.x_width):
    plt.plot(np.zeros(grid_map.x_width)+0.5+i, np.linspace(0, grid_map.y_width, grid_map.y_width)-0.5, 'white')
for i in range(grid_map.y_width):
    plt.plot(np.linspace(0, grid_map.x_width, grid_map.x_width)-0.5, np.zeros(grid_map.y_width)+0.5+i, 'white') 
plt.plot(np.array(path_x)-0.5, np.array(path_y)-0.5, 'cyan', linewidth=3.0)    
plt.plot(np.array(ps_path_x)-0.5, np.array(ps_path_y)-0.5, 'pink', linewidth=3.0)
plt.scatter(2-0.5, 2-0.5)
plt.scatter(12-0.5, 12-0.5)
plt.show()
```

```python
import heapq
import math
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
#set up matplotlib
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
plt.ion()
plt.figure(figsize=(12, 12))
MAX_INT = 9999.0

class GridMap:
    def __init__(self, ox, oy, resolution):
        self.min_x = None
        self.min_y = None
        self.max_x = None
        self.max_y = None
        self.x_width = None
        self.y_width = None
        self.obstacle_map = None
        
        self.resolution = resolution
        self.calc_obstacle_map(ox, oy)
        
    def calc_obstacle_map(self, ox, oy):

        self.min_x = round(min(ox))
        self.min_y = round(min(oy))
        self.max_x = round(max(ox))
        self.max_y = round(max(oy))
        print("min_x:", self.min_x)
        print("min_y:", self.min_y)
        print("max_x:", self.max_x)
        print("max_y:", self.max_y)

        self.x_width = round((self.max_x - self.min_x) / self.resolution)+1
        self.y_width = round((self.max_y - self.min_y) / self.resolution)+1
        print("x_width:", self.x_width)
        print("y_width:", self.y_width)

        # obstacle map generation
        self.obstacle_map = np.zeros((self.x_width, self.y_width))
        for x, y in zip(ox, oy):
            ix, iy = self.calc_xy_index(x, self.min_x), self.calc_xy_index(y, self.min_y)
            if 0<=ix<self.x_width and 0<=iy<self.y_width:
                self.obstacle_map[ix, iy] = 10.0  
        

    def calc_position(self, index, minp):
        pos = index * self.resolution + minp
        return pos
    
    def get_cx1x2_by_two_nodes(self, x1, y1, x2, y2):
        if x1 == x2 and y1 != y2:
            iyc = iyb = min(y1, y2)
            ixc = max(x1 - 1, 0)
            ixb = min(x1, self.x_width-1)
            return min(self.get_grid_cost(ixc, iyc), self.get_grid_cost(ixb, iyb))*self.resolution
            
        if x1 != x2 and y1 == y2:
            ixc = ixb = min(x1, x2)
            iyc = max(y1 - 1, 0)
            iyb = min(y1, self.y_width-1)
            return min(self.get_grid_cost(ixc, iyc), self.get_grid_cost(ixb, iyb))*self.resolution 
            
        if x1 != x2 and y1 != y2:
            ixc = min(x1, x2)
            iyc = min(y1, y2)
            return self.get_grid_cost(ixc, iyc)*self.resolution*1.414
            
    def get_grid_cost(self, ix, iy):
        if self.verify_node(ix, iy) == False: return 1.0
        if abs(self.obstacle_map[ix, iy]-10.0) < 0.01:
            return MAX_INT  
        else:
            return 1.0            

    def calc_xy_index(self, position, minp):
        return round((position - minp) / self.resolution)
    
    def calc_unique_index(self, ix, iy):
        return iy*self.x_width + ix

    def verify_node(self, ix, iy):
        if ix >= self.x_width or ix < 0 or iy >= self.y_width or iy < 0:
            return False      
        return True

sx = -5.0  # [m]
sy = -5.0  # [m]
gx = 50.0  # [m]
gy = 50.0  # [m]
grid_size = 5.0  # [m]
robot_radius = 1.0  # [m]

# set obstacle positions
ox, oy = [], []
for i in range(-10, 60):
    ox.append(i)
    oy.append(-10.0)
for i in range(-10, 60):
    ox.append(60.0)
    oy.append(i)
for i in range(-10, 61):
    ox.append(i)
    oy.append(60.0)
for i in range(-10, 61):
    ox.append(-10.0)
    oy.append(i)
# for i in range(-10, 40):
#     ox.append(20.0)
#     oy.append(i)
for i in range(20, 40):
    ox.append(40.0)
    oy.append(60.0 - i)    
    
for i in range(20, 40):
    oy.append(40.0)
    ox.append(i)    

grid_map = GridMap(ox, oy, grid_size)
grid_map.obstacle_map[11, 5] = 10.0
grid_map.obstacle_map[5, 11] = 10.0


class Node:
    def __init__(self, ix, iy, C=0.0, P=None):
        self.ix = ix
        self.iy = iy
        self.C = C
        self.P = P
        
action_set = {'UP':[0, 1], 'DOWN':[0, -1], 'LEFT':[-1, 0], 'RIGHT':[1, 0], 'UL':[-1, 1], 'UR':[1, 1], 'DL':[-1, -1], 'DR':[1, -1]}

class ThetaStar(object):
    def __init__(self, grid_map, start_node, end_node):
        self.start_node = start_node
        self.end_node = end_node
        self.grid_map = grid_map
        self.start_uid = self.grid_map.calc_unique_index(start_node.ix, start_node.iy)
        self.end_uid = self.grid_map.calc_unique_index(end_node.ix, end_node.iy)
        self.Closed_set = {}
        self.Open_set = {}
        self.Q = []
        self.Open_set[self.start_uid] = start_node
        heapq.heappush(self.Q, (start_node.C+self.get_heuristic(start_node), 
                           self.grid_map.calc_unique_index(start_node.ix, start_node.iy))) 
        
    def get_heuristic(self, node):
        dx = node.ix - self.end_node.ix
        dy = node.iy - self.end_node.iy
        return np.sqrt(dx*dx+dy*dy)*self.grid_map.resolution  
        

    def get_neighbors(self, current):
        for a in action_set:
            ix, iy = current.ix+action_set[a][0], current.iy+action_set[a][1]
            if self.grid_map.verify_node(ix, iy):
                cx1x2 = self.grid_map.get_cx1x2_by_two_nodes(current.ix, current.iy, ix, iy)                   
                yield (ix, iy, cx1x2)  
                
    def compute_shortest_path(self):
        while True:
            if not self.Open_set:
                print("A star Algorithm Finished.")
                return [], [], -1

            _, c_id = heapq.heappop(self.Q)

            if c_id in self.Open_set:
                current = self.Open_set.pop(c_id)
                self.Closed_set[c_id] = current
                if current.ix == end_node.ix and current.iy == end_node.iy:
                    return self.ShortestPathFinder()
                
                f_id = current.P
                father = Node(0, 0)
                if f_id: father = self.Closed_set[current.P]                    
                for ix, iy, cx1x2 in self.get_neighbors(current):
                    neighbor_uid = self.grid_map.calc_unique_index(ix, iy)
                    if neighbor_uid in self.Closed_set: continue
                    if neighbor_uid not in self.Open_set:
                        self.Open_set[neighbor_uid] = Node(ix, iy, current.C+cx1x2, c_id)
                        heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid)) 
                    fcx1x2 = 0.0
                    if f_id: fcx1x2 = self.LineOfSight(father.ix, father.iy, ix, iy)
                    
                    if f_id and abs(fcx1x2-MAX_INT) > 0.1 and self.Open_set[neighbor_uid].C > father.C + fcx1x2:
                        self.Open_set[neighbor_uid].C = father.C + fcx1x2
                        self.Open_set[neighbor_uid].P = current.P
                        heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid))   
                    elif self.Open_set[neighbor_uid].C > current.C + cx1x2:
                        self.Open_set[neighbor_uid].C = current.C + cx1x2
                        self.Open_set[neighbor_uid].P = c_id 
                        heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid))   
                        
                    
    def ShortestPathFinder(self):
        SPath_x = [self.Closed_set[self.end_uid].ix]
        SPath_y = [self.Closed_set[self.end_uid].iy]      
        SPath_dist = self.Closed_set[self.end_uid].C*self.grid_map.resolution
        v_index = self.Closed_set[self.end_uid].P
        while v_index: 
            SPath_x.append(self.Closed_set[v_index].ix)
            SPath_y.append(self.Closed_set[v_index].iy)
            v_index = self.Closed_set[v_index].P
        return list(reversed(SPath_x)), list(reversed(SPath_y)), SPath_dist   
    
    def LineOfSight(self, ix1, iy1, ix2, iy2):
        xL = ix2 - ix1
        yL = iy2 - iy1
        sign_x, sign_y = 1, 1
        if xL < 0: xL, sign_x = -xL, -1
        if yL < 0: yL, sign_y = -yL, -1
        f = 0
        cx1x2 = np.sqrt(xL*xL+yL*yL)*self.grid_map.resolution  
        if xL >= yL:
            while ix1 != ix2:
                f += yL
                if f >= xL:
                    if self.grid_map.get_grid_cost(ix1+((sign_x-1)//2), iy1+((sign_y-1)//2)) > 1.0:
                        return MAX_INT
                    iy1 += sign_y
                    f -= xL
                if f != 0 and self.grid_map.get_grid_cost(ix1+((sign_x-1)//2), iy1+((sign_y-1)//2)) > 1.0:
                    return MAX_INT
                if (yL == 0 and 
                   self.grid_map.get_grid_cost(ix1+((sign_x-1)//2), iy1) > 1.0 and
                   self.grid_map.get_grid_cost(ix1+((sign_x-1)//2), iy1-1)>1.0):
                    return MAX_INT
                ix1 += sign_x
        else:
            while iy1 != iy2:
                f += xL
                if f >= yL:
                    if self.grid_map.get_grid_cost(ix1+((sign_x-1)//2), iy1+((sign_y-1)//2)) > 1.0:
                        return MAX_INT
                    ix1 += sign_x
                    f -= yL
                if f != 0 and self.grid_map.get_grid_cost(ix1+((sign_x-1)//2), iy1+((sign_y-1)//2)) > 1.0:
                    return MAX_INT
                if (xL == 0 and 
                   self.grid_map.get_grid_cost(ix1, iy1+((sign_y-1)//2)) > 1.0 and
                   self.grid_map.get_grid_cost(ix1-1, iy1+((sign_y-1)//2))>1.0):
                    return MAX_INT
                iy1 += sign_y
        return cx1x2
                    
        
            
    def PostSmoothPath(self, path_x, path_y):
        k = 0
        ps_path_x, ps_path_y = [path_x[0]], [path_y[0]]
        for i in range(1, len(path_x)-1):
            if abs(self.LineOfSight(ps_path_x[k], ps_path_y[k], path_x[i+1], path_y[i+1])-MAX_INT) < 0.1:
                k += 1
                ps_path_x.append(path_x[i])
                ps_path_y.append(path_y[i])
        
        ps_path_x.append(path_x[-1])
        ps_path_y.append(path_y[-1])
        return ps_path_x, ps_path_y       

start_node = Node(2, 2)
end_node = Node(12, 12)
plt.imshow(grid_map.obstacle_map)
import time
t0 = time.time()
planner = ThetaStar(grid_map, start_node, end_node)
path_x, path_y, path_dist = planner.compute_shortest_path()
# ps_path_x, ps_path_y = planner.PostSmoothPath(path_x, path_y)
print("time cost is:", time.time()-t0)
print(path_x, path_y)
grid_map.obstacle_map[end_node.ix, end_node.iy] = 4
for ix, iy in zip(path_x, path_y):
    grid_map.obstacle_map[ix, iy] = 4  

plt.scatter(2-0.5, 2-0.5, color='green', linewidth=12.0)
plt.scatter(12-0.5, 12-0.5, color='red', linewidth=12.0)
# plt grid
for i in range(grid_map.x_width):
    plt.plot(np.zeros(grid_map.x_width)+0.5+i, np.linspace(0, grid_map.y_width, grid_map.y_width)-0.5, 'white')
for i in range(grid_map.y_width):
    plt.plot(np.linspace(0, grid_map.x_width, grid_map.x_width)-0.5, np.zeros(grid_map.y_width)+0.5+i, 'white') 
plt.plot(np.array(path_y)-0.5, np.array(path_x)-0.5, 'cyan', linewidth=3.0)    
plt.scatter(2-0.5, 2-0.5)
plt.scatter(12-0.5, 12-0.5)
plt.scatter(10-0.5,  11-0.5)
plt.scatter(11-0.5,  5-0.5)
plt.show()
```

```python
import math
def LineOfSight(ix1, iy1, ix2, iy2):            
    xL = abs(ix2 - ix1)
    yL = abs(iy2 - iy1)
    cx1x2 = np.sqrt(xL*xL+yL*yL)*grid_map.resolution  
    sign_x, sign_y = 1, 1
    if ix2 < ix1: sign_x = -1
    if iy2 < iy1: sign_y = -1
    print(sign_x, sign_y)

    if ix1 == ix2 and iy1 != iy2:
        for i in range(yL):
            iyc = iyb = min(iy1 + sign_y*i, iy1 + sign_y*(i+1))
            ixc = max(ix1 - 1, 0)
            ixb = min(ix1, grid_map.x_width-1)
            if min(grid_map.get_grid_cost(ixc, iyc), grid_map.get_grid_cost(ixb, iyb)) > 1.0:
                return MAX_INT
        return cx1x2

    if ix1 != ix2 and iy1 == iy2:
        for i in range(xL):
            ixc = ixb = min(ix1 + sign_x*i, ix1 + sign_x*(i+1))
            iyc = max(iy1 - 1, 0)
            iyb = min(iy1, grid_map.y_width-1)
            if min(grid_map.get_grid_cost(ixc, iyc), grid_map.get_grid_cost(ixb, iyb)) > 1.0:
                return MAX_INT
        return cx1x2
    
    tan_theta = yL/xL
    print(tan_theta)
    pre_y = iy1
    for i in range(xL+1):
        ixt = ix1 + sign_x*i
        iyt = iy1 + sign_y*math.floor(i*tan_theta)            
        if grid_map.get_grid_cost(ixt, iyt) > 1.0:
            return MAX_INT
        if i>0 and abs(iy1 + sign_y*i*tan_theta  - pre_y) > 1.0:
            print(abs(iy1 + sign_y*i*tan_theta  - pre_y), math.ceil(abs(iy1 + sign_y*i*tan_theta  - pre_y)))
            for k in range(1, math.ceil(abs(iy1 + sign_y*i*tan_theta  - pre_y))):                
                if grid_map.get_grid_cost(ixt-sign_x*1, pre_y+sign_y*k) > 1.0:
                    return MAX_INT    
        pre_y = iyt
            
    return cx1x2    
print(LineOfSight(2, 2, 6, 11))
```

```python
import heapq
import math
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
#set up matplotlib
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
plt.ion()
plt.figure(figsize=(12, 12))
MAX_INT = 9999.0

class GridMap:
    def __init__(self, ox, oy, resolution):
        self.min_x = None
        self.min_y = None
        self.max_x = None
        self.max_y = None
        self.x_width = None
        self.y_width = None
        self.obstacle_map = None
        
        self.resolution = resolution
        self.calc_obstacle_map(ox, oy)
        
    def calc_obstacle_map(self, ox, oy):

        self.min_x = round(min(ox))
        self.min_y = round(min(oy))
        self.max_x = round(max(ox))
        self.max_y = round(max(oy))
        print("min_x:", self.min_x)
        print("min_y:", self.min_y)
        print("max_x:", self.max_x)
        print("max_y:", self.max_y)

        self.x_width = round((self.max_x - self.min_x) / self.resolution)+1
        self.y_width = round((self.max_y - self.min_y) / self.resolution)+1
        print("x_width:", self.x_width)
        print("y_width:", self.y_width)

        # obstacle map generation
        self.obstacle_map = np.zeros((self.x_width, self.y_width))
        for x, y in zip(ox, oy):
            ix, iy = self.calc_xy_index(x, self.min_x), self.calc_xy_index(y, self.min_y)
            if 0<=ix<self.x_width and 0<=iy<self.y_width:
                self.obstacle_map[ix, iy] = 10.0  
        

    def calc_position(self, index, minp):
        pos = index * self.resolution + minp
        return pos
    
    def get_cx1x2_by_two_nodes(self, x1, y1, x2, y2):
        if x1 == x2 and y1 != y2:
            iyc = iyb = min(y1, y2)
            ixc = max(x1 - 1, 0)
            ixb = min(x1, self.x_width-1)
            return min(self.get_grid_cost(ixc, iyc), self.get_grid_cost(ixb, iyb))*self.resolution
            
        if x1 != x2 and y1 == y2:
            ixc = ixb = min(x1, x2)
            iyc = max(y1 - 1, 0)
            iyb = min(y1, self.y_width-1)
            return min(self.get_grid_cost(ixc, iyc), self.get_grid_cost(ixb, iyb))*self.resolution 
            
        if x1 != x2 and y1 != y2:
            ixc = min(x1, x2)
            iyc = min(y1, y2)
            return self.get_grid_cost(ixc, iyc)*self.resolution*1.414
            
    def get_grid_cost(self, ix, iy):
        if self.verify_node(ix, iy) == False: return 1.0
        if abs(self.obstacle_map[ix, iy]-10.0) < 0.01:
            return MAX_INT  
        else:
            return 1.0            

    def calc_xy_index(self, position, minp):
        return round((position - minp) / self.resolution)
    
    def calc_unique_index(self, ix, iy):
        return iy*self.x_width + ix

    def verify_node(self, ix, iy):
        if ix >= self.x_width or ix < 0 or iy >= self.y_width or iy < 0:
            return False      
        return True

sx = -5.0  # [m]
sy = -5.0  # [m]
gx = 50.0  # [m]
gy = 50.0  # [m]
grid_size = 5.0  # [m]
robot_radius = 1.0  # [m]

# set obstacle positions
ox, oy = [], []
for i in range(-10, 60):
    ox.append(i)
    oy.append(-10.0)
for i in range(-10, 60):
    ox.append(60.0)
    oy.append(i)
for i in range(-10, 61):
    ox.append(i)
    oy.append(60.0)
for i in range(-10, 61):
    ox.append(-10.0)
    oy.append(i)
# for i in range(-10, 40):
#     ox.append(20.0)
#     oy.append(i)
for i in range(20, 40):
    ox.append(40.0)
    oy.append(60.0 - i)    
    
for i in range(20, 40):
    oy.append(40.0)
    ox.append(i)    

grid_map = GridMap(ox, oy, grid_size)
grid_map.obstacle_map[11, 5] = 10.0
grid_map.obstacle_map[5, 11] = 10.0


class Node:
    def __init__(self, ix, iy, C=0.0, P=None):
        self.ix = ix
        self.iy = iy
        self.lb = -MAX_INT
        self.ub = MAX_INT
        self.C = C
        self.P = P
        
action_set = {'UP':[0, 1], 'DOWN':[0, -1], 'LEFT':[-1, 0], 'RIGHT':[1, 0], 'UL':[-1, 1], 'UR':[1, 1], 'DL':[-1, -1], 'DR':[1, -1]}

class ThetaStar(object):
    def __init__(self, grid_map, start_node, end_node):
        self.start_node = start_node
        self.end_node = end_node
        self.grid_map = grid_map
        self.start_uid = self.grid_map.calc_unique_index(start_node.ix, start_node.iy)
        self.end_uid = self.grid_map.calc_unique_index(end_node.ix, end_node.iy)
        self.Closed_set = {}
        self.Open_set = {}
        self.Q = []
        self.Open_set[self.start_uid] = start_node
        heapq.heappush(self.Q, (start_node.C+self.get_heuristic(start_node), 
                           self.grid_map.calc_unique_index(start_node.ix, start_node.iy))) 
        
    def get_heuristic(self, node):
        dx = node.ix - self.end_node.ix
        dy = node.iy - self.end_node.iy
        return np.sqrt(dx*dx+dy*dy)*self.grid_map.resolution  
        

    def get_neighbors(self, current):
        for a in action_set:
            ix, iy = current.ix+action_set[a][0], current.iy+action_set[a][1]
            if self.grid_map.verify_node(ix, iy):
                cx1x2 = self.grid_map.get_cx1x2_by_two_nodes(current.ix, current.iy, ix, iy)                   
                yield (ix, iy, cx1x2)  
                
    def compute_shortest_path(self):
        while True:
            if not self.Open_set:
                print("A star Algorithm Finished.")
                return [], [], -1

            _, c_id = heapq.heappop(self.Q)

            if c_id in self.Open_set:
                current = self.Open_set.pop(c_id)
                self.Closed_set[c_id] = current
                if current.ix == end_node.ix and current.iy == end_node.iy:
                    return self.ShortestPathFinder()
                
                f_id = current.P
                father = Node(0, 0)
                if f_id: father = self.Closed_set[current.P]                    
                for ix, iy, cx1x2 in self.get_neighbors(current):
                    neighbor_uid = self.grid_map.calc_unique_index(ix, iy)
                    if neighbor_uid in self.Closed_set: continue
                    if neighbor_uid not in self.Open_set:
                        self.Open_set[neighbor_uid] = Node(ix, iy, current.C+cx1x2, c_id)
                        heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid)) 
                    fcx1x2 = 0.0
                    if f_id: fcx1x2 = self.LineOfSight(father.ix, father.iy, ix, iy)
                    
                    if f_id and abs(fcx1x2-MAX_INT) > 0.1 and self.Open_set[neighbor_uid].C > father.C + fcx1x2:
                        self.Open_set[neighbor_uid].C = father.C + fcx1x2
                        self.Open_set[neighbor_uid].P = current.P
                        heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid))   
                    elif self.Open_set[neighbor_uid].C > current.C + cx1x2:
                        self.Open_set[neighbor_uid].C = current.C + cx1x2
                        self.Open_set[neighbor_uid].P = c_id 
                        heapq.heappush(self.Q, (self.Open_set[neighbor_uid].C+self.get_heuristic(self.Open_set[neighbor_uid]), neighbor_uid))   
                        
                    
    def ShortestPathFinder(self):
        SPath_x = [self.Closed_set[self.end_uid].ix]
        SPath_y = [self.Closed_set[self.end_uid].iy]      
        SPath_dist = self.Closed_set[self.end_uid].C*self.grid_map.resolution
        v_index = self.Closed_set[self.end_uid].P
        while v_index: 
            SPath_x.append(self.Closed_set[v_index].ix)
            SPath_y.append(self.Closed_set[v_index].iy)
            v_index = self.Closed_set[v_index].P
        return list(reversed(SPath_x)), list(reversed(SPath_y)), SPath_dist   
    
                    
        
            
    def PostSmoothPath(self, path_x, path_y):
        k = 0
        ps_path_x, ps_path_y = [path_x[0]], [path_y[0]]
        for i in range(1, len(path_x)-1):
            if abs(self.LineOfSight(ps_path_x[k], ps_path_y[k], path_x[i+1], path_y[i+1])-MAX_INT) < 0.1:
                k += 1
                ps_path_x.append(path_x[i])
                ps_path_y.append(path_y[i])
        
        ps_path_x.append(path_x[-1])
        ps_path_y.append(path_y[-1])
        return ps_path_x, ps_path_y       

start_node = Node(2, 2)
end_node = Node(12, 12)
plt.imshow(grid_map.obstacle_map)
import time
t0 = time.time()
planner = ThetaStar(grid_map, start_node, end_node)
path_x, path_y, path_dist = planner.compute_shortest_path()
# ps_path_x, ps_path_y = planner.PostSmoothPath(path_x, path_y)
print("time cost is:", time.time()-t0)
print(path_x, path_y)
grid_map.obstacle_map[end_node.ix, end_node.iy] = 4
for ix, iy in zip(path_x, path_y):
    grid_map.obstacle_map[ix, iy] = 4  

plt.scatter(2-0.5, 2-0.5, color='green', linewidth=12.0)
plt.scatter(12-0.5, 12-0.5, color='red', linewidth=12.0)
# plt grid
for i in range(grid_map.x_width):
    plt.plot(np.zeros(grid_map.x_width)+0.5+i, np.linspace(0, grid_map.y_width, grid_map.y_width)-0.5, 'white')
for i in range(grid_map.y_width):
    plt.plot(np.linspace(0, grid_map.x_width, grid_map.x_width)-0.5, np.zeros(grid_map.y_width)+0.5+i, 'white') 
plt.plot(np.array(path_y)-0.5, np.array(path_x)-0.5, 'cyan', linewidth=3.0)    
plt.scatter(2-0.5, 2-0.5)
plt.scatter(12-0.5, 12-0.5)
plt.scatter(10-0.5,  11-0.5)
plt.scatter(11-0.5,  5-0.5)
plt.show()
```
